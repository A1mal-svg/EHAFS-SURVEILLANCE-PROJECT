# EHAFS — visualize attention on RWF-2000
Loads a checkpoint and overlays per-frame attention weights on sampled frames.

```python
import sys, pathlib
sys.path.insert(0, str(pathlib.Path('..').resolve()))
import matplotlib.pyplot as plt
from ehafs.inference import EHAFSPredictor

predictor = EHAFSPredictor(checkpoint='../checkpoints/ehafs_best.pt')
result = predictor.predict('../data/RWF-2000/val/Fight/some_clip.avi')
print(result['label'], result['probs'])

frames, attn = result['frames'], result['attention']
fig, axes = plt.subplots(2, 8, figsize=(20, 6))
for i, ax in enumerate(axes.flat):
    ax.imshow(frames[i]); ax.set_title(f't={i+1} α={attn[i]:.3f}')
    ax.axis('off')
plt.tight_layout(); plt.show()
```
